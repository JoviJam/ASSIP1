```python
import pandas as pd
```

```python
penguins_url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/c19a904462482430170bfe2c718775ddb7dbb885/inst/extdata/penguins.csv"
penguins_df = pd.read_csv(penguins_url)
penguins_df.head()
```

## Exercise 2

Pick one numerical column and make an appropiate plot. Comment your findings.

Optional: Make an innappropiate plot.

```python
import matplotlib.pyplot as plt
plt.plot(penguins_df['bill_length_mm'], 'o', color='red')
plt.title('Bill Length')
plt.xlabel('Index')
plt.ylabel('Length')
plt.show()
```

This shows the lengths of all the bills in order throughout the dataset.

## Exercise 3

Pick one categorical column and make an appropiate plot. Comment your findings.

Optional: Make an innappropiate plot.

```python
penguins_df.dropna()
species_count = penguins_df['species'].value_counts()
colors = {'Adelie': 'blue', 'Chinstrap': 'green', 'Gentoo': 'red'}
plt.bar(species_count.index, species_count.values, color = [colors[species] for species in species_count.index])
plt.title('Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

## Exercise 4:

Pick two columns and make an appropiate plot. Feel free to try as many different combinations.

```python
plt.plot(penguins_df['bill_length_mm'], penguins_df['bill_depth_mm'], 'o', color='yellow')
plt.title('Bill Length vs Bill Depth')
plt.xlabel('Length')
plt.ylabel('Depth')
plt.show()
```
