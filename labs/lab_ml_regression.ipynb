```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/aoguedao/gmu-assip-2024/main/data/Diamond%20Prices%202022.csv"
diamonds_df = pd.read_csv(url)
diamonds_df.head()
```

```python
# Remove the column "index"
diamonds_df = diamonds_df.drop(columns=["index"])
diamonds_df
```

```python
# Plot price vs carat
x = diamonds_df["price"]
y = diamonds_df["carat"]
plt.scatter(x, y, color = 'green')
plt.title("Price vs Carat")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```

```python
# Plot price vs clarity
x = diamonds_df["price"]
y = diamonds_df["clarity"]
plt.scatter(x, y)
plt.title("Price vs Clarity")
plt.xlabel("Price")
plt.ylabel("Clarity")
plt.show()
```

```python
# Plot price vs table
x = diamonds_df["price"]
y = diamonds_df["table"]
plt.scatter(x, y, color = 'purple')
plt.title("Price vs Table")
plt.xlabel("Price")
plt.ylabel("Table")
plt.show()
```

Can you see any relationship?
